# **Encoding And Types of Encoding**

# Encoding:
       Encoding in data analysis means converting categorical or text data into numerical form so a computer or machine learning model can understand and use it.

Example:

```
Gender: Male, Female
```

After encoding:

```
Male = 0
Female = 1
```

Because ML algorithms work with numbers, not text.

---

## **Types of Encoding**

### 1 Label Encoding

* Converts each category into a unique number.
* Best for ordinal data (data with order).

Example:

```
Low = 0
Medium = 1
High = 2
```

---

### 2 One-Hot Encoding

* Creates separate columns for each category.
* Used for nominal data (no order).

Example:

```
Color: Red, Blue, Green
```

Becomes:

```
Red  Blue  Green
1     0      0
0     1      0
0     0      1
```

---

### 3 Ordinal Encoding

* Similar to label encoding but used when categories have ranking.

Example:

```
Education: Primary < Secondary < Bachelor < Master
```

---

### 4 Binary Encoding

* Converts categories into binary numbers (0/1).
* Useful when there are many categories.

---

### 5 Frequency Encoding

* Categories replaced with how often they appear.

Example:

```
City:
Karachi = 50 times
Lahore = 30 times
```

---

### 6 Target Encoding

* Categories replaced with average of target variable.
* Used in machine learning models.

---

## In short

Encoding = transforming text/categorical data into numbers so models can understand it.
