In ML, you dont randomly choose normalization or standardization  it depends on the **distribution and nature of your data** and the **algorithm youre using**.

## 1) First understand the difference

**Normalization (MinMax scaling)**

* Scales data to a fixed range (usually 01)
* Formula:
  [
  x' = \frac{x - min(x)}{max(x) - min(x)}
  ]
* Keeps original distribution shape
* Sensitive to outliers

**Standardization (Z-score scaling)**

* Centers data around mean = 0 and std = 1
* Formula:
  [
  x' = \frac{x - \mu}{\sigma}
  ]
* Works well when data follows normal distribution
* Less affected by outliers

---

## 2) How to identify which one to use

### Use **Normalization** when:

* Data has **no normal (Gaussian) distribution**
* Features have **different ranges** (e.g., age 1100, salary 10k1M)
* Algorithm depends on **distance**

  * KNN
  * K-means
  * Neural Networks
  * Gradient Descent models
* Pixel values (0255)  normalize to 01

Example:

```
Height: 150190
Weight: 5095
Income: 20k300k
```

 Normalize

---

### Use **Standardization** when:

* Data is **normally distributed**
* Outliers exist
* Algorithms assume mean-centered data:

  * Linear Regression
  * Logistic Regression
  * SVM
  * PCA
  * LDA

Example:

```
Exam scores with bell curve
Financial returns
Scientific measurements
```

 Standardize

---

## 3) Practical way to decide (most important)

### Step 1  Check distribution

```python
import seaborn as sns
sns.histplot(df['feature'])
```

* Bell shape  Standardization
* Skewed  Normalization

### Step 2  Check algorithm

| Algorithm           | Scaling         |
| ------------------- | --------------- |
| KNN                 | Normalization   |
| K-means             | Normalization   |
| Neural Network      | Normalization   |
| Linear Regression   | Standardization |
| Logistic Regression | Standardization |
| SVM                 | Standardization |
| PCA                 | Standardization |

### Step 3  Check outliers

* Many outliers  Standardization
* No outliers  Normalization

---

## 4) How to apply in Python

### Normalization

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df_scaled = scaler.fit_transform(df)
```

### Standardization

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df_scaled = scaler.fit_transform(df)
```

---

## 5) Golden rule (interview answer)

* If data is **Gaussian  Standardization**
* If data range matters  Normalization
* If using **distance-based models  Normalization**
* If using **statistical models  Standardization**

---

## 6) Real ML workflow

1. Do EDA
2. Check distribution
3. Check outliers
4. Select algorithm
5. Then choose scaling
