```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
df=pd.read_csv('House Prices.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
df.drop(columns='Date',inplace=True)
```

```python
df.dtypes
```

```python
df.duplicated().sum()
```

```python
plt.figure(figsize=(15,7))
sns.heatmap(df.corr(),annot=True)
plt.show()
```

```python
x=df.drop(columns='Price')
y=df['Price']
```

```python
st=StandardScaler()
st.fit(x)
x=pd.DataFrame(st.transform(x),columns=x.columns)
```

```python
x
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression, Lasso, Ridge
```

```python
lr=LinearRegression()
lr.fit(x_train,y_train)
lr.score(x_test,y_test)*100
```

```python
plt.figure(figsize=(22,10))
plt.bar(x.columns,lr.coef_)
plt.title('Linear Regession')
plt.xlabel('Columns')
plt.ylabel('Cofficient')
plt.show()
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
mse=mean_squared_error(y_test,lr.predict(x_test))
mae=mean_absolute_error(y_test,lr.predict(x_test))
print(mse)
print(mae)
print(np.sqrt(mean_squared_error(y_test,lr.predict(x_test))))
```

```python
ls=Lasso(alpha=0.01)
ls.fit(x_train,y_train)
ls.score(x_test,y_test)*100
```

```python
plt.figure(figsize=(22,10))
plt.bar(x.columns,ls.coef_)
plt.title('Lasso')
plt.xlabel('Columns')
plt.ylabel('Cofficient')
plt.show()
```

```python
mse=mean_squared_error(y_test,ls.predict(x_test))
mae=mean_absolute_error(y_test,ls.predict(x_test))
print(mse)
print(mae)
print(np.sqrt(mean_squared_error(y_test,ls.predict(x_test))))
```

```python
ri=Ridge(alpha=10)
ri.fit(x_train,y_train)
ri.score(x_test,y_test)*100
```

```python
plt.figure(figsize=(22,10))
plt.bar(x.columns,ri.coef_)
plt.title('Ridge')
plt.xlabel('Columns')
plt.ylabel('Cofficient')
plt.show()
```

```python
mse=mean_squared_error(y_test,ri.predict(x_test))
mae=mean_absolute_error(y_test,ri.predict(x_test))
print(mse)
print(mae)
print(np.sqrt(mean_squared_error(y_test,ri.predict(x_test))))
```
